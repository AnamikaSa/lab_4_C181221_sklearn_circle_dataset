```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error
from tqdm import tqdm_notebook 
from sklearn.preprocessing import OneHotEncoder
from sklearn.datasets import make_blobs
```

```python
my_cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ["pink","yellow","green"])
```

```python
np.random.seed(0)
```

```python
from sklearn.datasets import make_moons, make_circles
```

```python
data, labels = make_moons(n_samples=6000, random_state=0, noise=0.15)
print(data.shape, labels.shape)
```

```python
plt.scatter(data[:,0], data[:,1], c=labels, cmap=my_cmap)
plt.show()
```

```python
data, labels = make_circles(n_samples=1000, random_state=0, noise=0.2, factor=0.3)
print(data.shape, labels.shape)
```

```python
plt.scatter(data[:,0], data[:,1], c=labels, cmap=my_cmap)
plt.show()
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(data, labels, stratify=labels, random_state=0)
print(X_train.shape, X_val.shape)
```
